```python
import os.path
import csv
import sqlite3
import pandas as pd
```

```python
BASE_DIR = os.path.dirname(os.path.abspath("/Users/zack/Desktop/part_db/part_granted.db"))
db_path = os.path.join(BASE_DIR, "part_granted.db")
```

```python
def pandify_from_sql():
    with sqlite3.connect(db_path) as db:
        c = db.cursor()
        query = c.execute("SELECT * From patents")
        cols = [column[0] for column in query.description]
        results= pd.DataFrame.from_records(data = query.fetchall(), columns = cols)
        return results
```

```python
dfsql = pandify_from_sql()
```

```python
dfsql['app_number'] = dfsql['app_number'].astype(str)
dfsql['app_number'] = dfsql['app_number'].str.rstrip()
dfsql['app_number'] = dfsql['app_number'].str[3:]
dfsql['app_number'] = dfsql['app_number'].astype(int)
```
